Tempo vs. Popularity by Genre:
○ (a) Use group-by-analysis to calculate the average popularity for different tempo
ranges across genres.
○ (b) Visualize this with a line plot showing tempo ranges on the x-axis and
popularity on the y-axis, segmented by genre.

```python
#import necessary packages
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
#handle null values
```

```python
#read csv into df
df=pd.read_csv("spotify_songs.csv")

#separate tempo into bins 
bins = [0, 25, 50, 75, 100, 125, 150, 175, 200, 225, 250]
labels = ['0-25', '26-50', '51-75', '76-100', '101-125', '126-150', '151-175', '176-200', '201-225', '225-250']
df['tempo_range'] = pd.cut(df['tempo'], bins=bins, labels=labels)

#group by to show average popularity by genre and tempo range 
df.groupby(["tempo_range","playlist_genre"])["track_popularity"].mean()
```

```python
#visualization
sns.scatterplot(data=df, x='tempo', y='track_popularity', hue='playlist_genre')
```
